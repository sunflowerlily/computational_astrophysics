```python
import numpy as np
```

```python
data = np.loadtxt("planetary_stability.dat")
```

```python
t = data[:,0]
vel = data[:, 4]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot(t, vel)
```

```python
F = np.fft.rfft(vel)
```

```python
kfreq = np.fft.rfftfreq(len(t)) * len(t) / t.max()
```

```python
fig, ax = plt.subplots()
ax.plot(kfreq, np.abs(F)**2 * 2 / len(t))
ax.set_xscale("log")
ax.set_yscale("log")
ax.grid()
```
